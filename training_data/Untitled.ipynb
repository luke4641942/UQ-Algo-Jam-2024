```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Thrifted Jeans_price_history.csv", index_col="Day", usecols= ["Day", "Price"], parse_dates=True)
df.head()
```

```python
df["Price"].plot(figsize=(10,6))
plt.title("Thrifted Jeans Price")
plt.xlabel("Day")
plt.ylabel("Price")
```

```python
# Calculate moving averages
df["12 Day MA"] = df["Price"].rolling(window=12).mean()
df["3 Day MA"] = df["Price"].rolling(window=3).mean()

# Plot the original price, 30-day MA, and 5-day MA
df["Price"].plot(figsize=(10, 6), label="Price")
df["12 Day MA"].plot(label="12 Day Moving Average", color="green")
df["3 Day MA"].plot(label="3 Day Moving Average", color="orange")

# Add title and labels
plt.title("Thrifted Jeans Price")
plt.xlabel("Day")
plt.ylabel("Price")

# Add a legend to distinguish the lines
plt.legend()

# Show the plot
plt.show()
```

```python
# Create a new DataFrame with the desired columns
df_with_moving_averages = df[["Price"]].copy()

# Add the 5-day and 30-day moving averages
df_with_moving_averages["5 Day MA"] = df["Price"].rolling(window=3).mean()
df_with_moving_averages["30 Day MA"] = df["Price"].rolling(window=12).mean()

# Reset the index to include the "Day" column explicitly if it's an index
df_with_moving_averages.reset_index(inplace=True)

# Display the first few rows of the table
df_with_moving_averages[30:70]
```
